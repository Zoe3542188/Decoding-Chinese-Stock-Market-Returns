```python
import numpy as np
from scipy.stats import kurtosis,skew
from openpyxl import load_workbook
```

```python
return_values=[]
workbook=load_workbook('4183returns.xlsx')
sheets = workbook.get_sheet_names()
booksheet = workbook.get_sheet_by_name(sheets[0]) 
rows = booksheet.rows
columns=1
for row in rows:  
    line = [col.value for col in row]
for i in range(1,4184):
    value=booksheet.cell(row=i,column=1).value
    return_values.append(value)
length=len(return_values)
print(length)
```

```python
print(return_values)
```

```python
std=np.std(return_values)
mean=np.mean(return_values)
print('- - - - - -Descriptive Statistics - - - - - -')
print('   Mean                 ',mean)
print(' ')
print('   Standard deviation   ',std)
print(' ')
print('   Skewness             ',skew(return_values))
print(' ')
print('   Kurtosis             ',kurtosis(return_values))
```
